```python
'''to plot the distribution of prices '''
from scipy.stats import binom
import numpy as np
rv = binom(10, 0.7)

x = np.arange(-1, 12, 1)
y = rv.pmf(x)

plt.bar(x-0.2, y, width=0.4)

plt.title("binomial distribution")
plt.xlabel("RV")
plt.ylabel("P(X=x)")
plt.show()
```

```python
from scipy.stats import geom
rv = geom(0.45)

x = np.arange(-1, 15, 1)
y = rv.pmf(x)

plt.bar(x-0.2, y, width=0.4)

plt.ylim([0, 0.5])
plt.title("geometric distribution")
plt.xlabel("RV")
plt.ylabel("P(X=x)")
plt.show()
```

```python
'''calculate the median of prices
calc the center of the prices
'''
from pymongo import MongoClient                                                            
from bson.objectid import ObjectId
from statistics import median
                                                                                           
prods = MongoClient('192.168.1.202')['etl-dev'].products                                   
pitem = MongoClient('192.168.1.202')['etl-dev'].product_items                              
                                                                                          
                        
def find_median_of_prices(prod_col, item_col , key): 
    
    p = prod_col.find_one({'_id':ObjectId("53bce7f88ddf879590b02c99")})    
    prices_taobao = []
    prices_others = []
    for ns in p[key]:

            
        query = item_col.find({'source':ns, 'product_id':{'$in': p[key][ns]}})        
        for q in query:      
            if ns == 'taobao':
                prices_taobao.append((q['unit_price'], q['source'], q['product_name']))      
            else:
                prices_others.append((q['unit_price'], q['source'], q['product_name']))                   
    prices_taobao = sorted(prices_taobao, key=lambda x:x[0], reverse=True)                                   
    prices_others = sorted(prices_others, key=lambda x:x[0], reverse=True)                                   
    
    pt = [i[0] for i in prices_taobao]
    po = [i[0] for i in prices_others]
        
    #print(median(p))
    return pt, po
    
pt, po = find_median_of_prices(prods, pitem, '_productsIds')
import matplotlib.pyplot as plt
plt.hist((pt,po))
plt.title("Gaussian Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
from pymongo import MongoClient                                                            
from bson.objectid import ObjectId
from statistics import median
import matplotlib.pyplot as plt  
import time
import numpy as np
import pandas as pd

prods = MongoClient('192.168.1.202')['etl-stage'].products                                   
pitem = MongoClient('192.168.1.202')['etl-stage'].product_items    

common_params = dict(bins=50, 
                     normed=True,
                     facecolor='gray')    

def find_median_of_prices(prod_col, item_col , key): 
    query = prods.find({'reviewSummary.count':{'$gt':30000}})
    for q in query:
       
        prices = []
        print(q['productName'])
        for k, v in q['_unitPrice'].items():
            prices.append(v[1])
        s = pd.Series(np.array(prices)).describe()
        one_quarter = s['25%']
        three_quarter = s['75%']
        IQR = three_quarter - one_quarter
        max_dist = 1.5 * IQR
        UpperLimit = three_quarter + max_dist
        LowerLimit = one_quarter - max_dist
        #----
        
        #----
        print('IQR = {}'.format(IQR))
        print('LowerLimit = {}'.format(LowerLimit))
        print('min value = {}'.format(min(prices)))
        print('UpperLimit = {}'.format(UpperLimit))
        print('max value = {}'.format(max(prices)))
        print()
        plt.hist(prices, **common_params)
        plt.title("Gaussian Histogram")
        plt.xlabel("Value")
        plt.ylabel("Frequency")
        plt.show()
        
find_median_of_prices(prods, pitem, '_productsIds')
```
