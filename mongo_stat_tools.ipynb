```python
'''
aggregate: each brand, each series, how many products exists
'''
from pymongo import MongoClient
from bson import SON
client = MongoClient()
db = client['caigen-development']
formula_products = db.formula_products
brands = formula_products.distinct('brandName')
for brand in brands:
    print('-----------%s----------'%brand)
    result = formula_products.aggregate([
        {'$match':{'brandName': brand}},
        {"$group": {"_id" :"$tag", "count": {"$sum" : 1}}},
        {"$sort" :SON([("count",-1), ("_id", -1)])}
    ])
    if result:
        for l in result['result']:
            print("%s : %s\n"%(l['_id'], l['count']))
    else:
        print('nothing')
```

```python

from pymongo import MongoClient
from bson import SON
def stat_store(host, db, collection, key):
    prod = MongoClient(host)[db]
    query = prod[collection].find({ key:{'$exists':1}}).limit(0)
    jd, yhd, tmall, amazon , taobao = 0,0,0,0,0
    for q in query:
        pids = q[key]
        jd += len(pids.get('jd',[]))
        yhd += len(pids.get('yhd',[]))
        tmall += len(pids.get('tmall',[]))
        taobao += len(pids.get('taobao',[]))
        amazon += len(pids.get('amazon',[]))
    print('jd',jd)
    print('yhd',yhd)
    print('tmall',tmall)    
    print('taobao',taobao)
    print('amazon',amazon)

#stat_store('192.168.1.202', 'caigen-development', 'products', 'productsIds')
stat_store('192.168.1.202', 'etl-dev', 'products', '_products_ids')
```

```python
'''cal total number of products'''
from pymongo import MongoClient
from bson import SON
def stat_store_total(host, db, collection, key):
    prod = MongoClient(host)[db]
    query = prod[collection].find({ key:{'$exists':1}}).limit(0)
    total = 0
    for q in query:
        total += sum([ len(q[key][store]) for store in q[key]])  
    print(total)

stat_store_total('192.168.1.21', 'caigen-development', 'products', 'products_ids')
stat_store_total('192.168.1.202', 'dev', 'products_max_review', 'productsIds')
```

```python
'''calculate the median of prices
taobao's price derived too large, so it will not be calculated
'''

from pymongo import MongoClient                                                            
from bson.objectid import ObjectId
from statistics import median
                                                                                           
prods = MongoClient('192.168.1.202')['etl-dev'].products                                   
pitem = MongoClient('192.168.1.202')['etl-dev'].product_items                              
                                                                                          
                        
def find_median_of_prices(prod_col, item_col , key):  
    p = prod_col.find_one({'_id':ObjectId("53bbc09f8ddf87a6aabe0016")})    
    prices = []                                                                                
    for ns in p[key]:
        if ns == 'taobao':
            continue
        query = item_col.find({'source':ns, 'product_id':{'$in': p[key][ns]}})        
        for q in query:                                                                        
            prices.append((q['unit_price'], q['source'], q['product_name']))                   
    prices = sorted(prices, key=lambda x:x[0], reverse=True)                                   
    for k in prices:
        print(k)
    p = [i[0] for i in prices]
    
    print(median(p))
    
find_median_of_prices(prods, pitem, '_productsIds')
```
